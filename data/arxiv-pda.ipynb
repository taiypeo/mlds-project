# Reference graph creation

Construct a graph of 2023 ARXIV papers from above, their references, and references of their references

```python
import requests
import json
import time
from datetime import datetime

def get_references(paper_ids):
    url = "https://api.semanticscholar.org/graph/v1/paper/batch"
    params = {"fields": "title,references.paperId,references.title,references"}
    body = {"ids": paper_ids}
    
    try:
        resp = requests.post(url, json=body, params=params)
    except:
        print(f"POST failed for ids {paper_ids}, retrying in 20s")
        time.sleep(20)
        return get_references(paper_ids)
        
    if resp.status_code != 200:
        print(f"Rate limit reached for ids {paper_ids}, retrying in 10s")
        time.sleep(10)
        return get_references(paper_ids)

    data = resp.text
    result = json.loads(data)
    for p in result:
        if p is None:
            print(f"Did not find one of the papers among {paper_ids}")
        else:
            assert p["paperId"] is not None
            if len(p["references"]) == 0:
                print(f"Empty references for paper {p['paperId']}")

    return result
```

```python
from tqdm import trange

MAX_PAPERS_PER_REQUEST = 100
paper_references = []

for start in trange(0, df_filtered.shape[0], MAX_PAPERS_PER_REQUEST):
    ids = ("ARXIV:" + df_filtered["id"].iloc[start:start+MAX_PAPERS_PER_REQUEST]).tolist()
    paper_references.extend(get_references(ids))
    time.sleep(0.1)
```

```python
unique_original_ids = set()
unique_ref_ids = set()
for paper in paper_references:
    if paper is None or paper["paperId"] is None:
        continue
    unique_original_ids.add(paper["paperId"])
    
    for ref in paper["references"]:
        if ref["paperId"] is None:
            continue
        
        unique_ref_ids.add(ref["paperId"])

len(unique_original_ids), len(unique_ref_ids)
```

```python
with open("unique_original_ids.json", "w") as file:
    json.dump(list(unique_original_ids), file)
```

```python
unique_ref_ids_list = list(unique_ref_ids)

print(len(paper_references))

for start in trange(0, len(unique_ref_ids_list), MAX_PAPERS_PER_REQUEST):
    ids = unique_ref_ids_list[start:start+MAX_PAPERS_PER_REQUEST]
    paper_references.extend(get_references(ids))
    time.sleep(0.05)

print(len(paper_references))
```

```python
paper_references_cleaned = [paper for paper in paper_references if paper is not None and paper["paperId"] is not None]

# Clean references with unknown IDs, as they are usually parsing junk
for paper in paper_references_cleaned:
    paper["references"] = [
        ref
        for ref in paper["references"]
        if ref["paperId"] is not None and (ref["paperId"] in unique_original_ids or ref["paperId"] in unique_ref_ids)
]

with open("paper_references_cleaned.json", "w") as file:
    json.dump(paper_references_cleaned, file)
```

```python
paper_id_to_references_ids = {}
for paper in paper_references_cleaned:
    paper_id_to_references_ids[paper["paperId"]] = [ref["paperId"] for ref in paper["references"]]
```

```python
unique_ids = set()
for paper_id, references in paper_id_to_references_ids.items():
    unique_ids.add(paper_id)
    unique_ids.update(references)

len(unique_ids)
```

```python
def get_paper_infos(paper_ids):    
    url = "https://api.semanticscholar.org/graph/v1/paper/batch"
    params = {"fields": "title,abstract,year"}
    body = {"ids": paper_ids}
    
    try:
        resp = requests.post(url, json=body, params=params)
    except:
        print(f"POST failed for ids {paper_ids}, retrying in 20s")
        time.sleep(20)
        return get_paper_infos(paper_ids)
    
    if resp.status_code != 200:
        print(f"Got code {resp.status_code} for ids {paper_ids}, retrying in 10s")
        time.sleep(10)
        return get_paper_infos(paper_ids)
    
    data = resp.text
    return json.loads(data)
```

```python
paper_info = []
unique_ids_list = list(unique_ids)
    
for start in trange(0, len(unique_ids_list), MAX_PAPERS_PER_REQUEST // 2):
    ids = unique_ids_list[start:start+MAX_PAPERS_PER_REQUEST // 2]
    paper_info.extend(get_paper_infos(ids))
    time.sleep(0.8)

len(paper_info)
```

```python
paper_info_filtered = {paper["paperId"]: paper for paper in paper_info if paper}
with open("papers.json", "w") as file:
    json.dump(paper_info_filtered, file)
```

```python
missing_ids = unique_ids - paper_info_filtered.keys()
len(missing_ids), len(unique_ids)
```

```python
all(k in paper_info_filtered for k in paper_id_to_references_ids.keys())
```

```python
paper_id_to_references_ids_filtered = {
    key: [r for r in ref if r not in missing_ids]
    for key, ref in paper_id_to_references_ids.items()
}

with open("graph.json", "w") as file:
    json.dump(paper_id_to_references_ids_filtered, file)
```
